```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import HeatMap, MarkerCluster, FastMarkerCluster, beautify_icon
from folium import Choropleth, Circle, Marker
import geopandas as gpd
```

```python
N_HQ_Hub_dis = pd.read_excel('data/N_HQ_Hub_dis.xlsx')
N_SBO_Hub_dis = pd.read_excel('data/N_SBO_Hub_dis.xlsx')
province_acv = pd.read_excel('data/N_Province_acv.xlsx')
DCP_MD = pd.read_excel('data/April_11_DCP Master Data.xlsx')
re_real = pd.read_excel('data/april_11/April_11_RD_cover_real_bp.xlsx')
dcp_cover_with_real = pd.read_excel('data/april_11/April_11_DCP_cover_real_bp(with DCP hub).xlsx')
dcp_cover_without_real = pd.read_excel('data/april_11/April_11_DCP_cover_real_bp(without DCP hub).xlsx')
```

```python
# 把两种hub 放到地图上
m =  folium.Map(location = [31.2304, 121.4737], zoom_start = 5)
tile_list = ['openstreetmap', 'cartodbpositron']

for each in tile_list:
    folium.TileLayer(each).add_to(m)

national_geo = gpd.read_file('./data/中华人民共和国.json')
folium.Choropleth(
    geo_data=national_geo,
    name="Blackpin Stores",
    data=province_acv,
    columns=['Province', 'Store Acv Score'],
    key_on="properties.name",
    fill_color="YlGn",
    fill_opacity=0.3,
    line_opacity=0.2,
    legend_name="National GaoDe Blackpin Stores",
).add_to(m)
```

```python
feature_group = folium.FeatureGroup(name="HQ平均半径", show=False)
feature_group1 = folium.FeatureGroup(name="HQ最大半径", show=False)
feature_group2 = folium.FeatureGroup(name="SBO平均半径", show=False)
feature_group3 = folium.FeatureGroup(name="SBO最大半径", show=False)
feature_group4 = folium.FeatureGroup(name="黑钉 Stores", show=False)
feature_group5 = folium.FeatureGroup(name="SBO BP Stores", show=False)
feature_group6 = folium.FeatureGroup(name="DCP BP Store", show=False)
feature_group11 = folium.FeatureGroup(name="DCP site", show=False)
feature_group12 = folium.FeatureGroup(name="DCP cover stores", show=False)
feature_group13 = folium.FeatureGroup(name="No DCP cover stores", show=False)
```
